```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
test_oligo = np.genfromtxt('Results/Oligo_2d_bi_eps1.0_nv200_sd0.07777777777777778.dat')
```

```python
fig, ax = plt.subplots()
timestep = test_oligo[:, 0]
mean_size = test_oligo[:, 1]
ax.plot(timestep, mean_size)
ax.set_xlabel('Time')
ax.set_ylabel('Mean Oligomer size')
```

```python
fig, ax = plt.subplots()

frame = -1
oligo_size_dist = test_oligo[frame, 2:]
x = np.arange(len(oligo_size_dist))
y = oligo_size_dist
ax.plot(x,y)

ax.set_title('Oligomer size distribution ')
ax.set_xlabel('Oligomer size')
```
